# Simple Sequential Chain

```python
from enum import verify
from langchain_core.prompts import PromptTemplate

llm = ChatGroq(model="openai/gpt-oss-20b", temperature=0.7, api_key = os.environ["GROQ_API_KEY"], verbose=True)


# ── Step 1: Generate a dataset description ──────────────────────────────────
prompt_1 = PromptTemplate(
    input_variables=["cuisine"],
    template="I want to open a restaurant for {cuisine} food, suggest some names for this."

)
chain_1 = prompt_1 | llm

# ── Step 2: Suggest ML models based on Step 1's output ──────────────────────
prompt_2 = PromptTemplate(
    input_variables=["restaurant_name"],
    template=(
        "Suggest some menu items for {restaurant_name}"

    )
)
chain_2 = prompt_2 | llm

# ── Compose: Output of chain_1 flows into chain_2 ───────────────────────────
# chain_1 returns a string → we wrap it into the dict chain_2 expects
sequential_chain = (
    chain_1
    | (lambda text: {"restaurant_name": text})  # bridge: str → dict
    | chain_2
)

# ── Run it ───────────────────────────────────────────────────────────────────
result = sequential_chain.invoke({'cuisine': "Indian"})
```

```python
from IPython.display import display, Markdown

display(Markdown(result.content))
```

# Simple Sequential Chain for Intermediate Output

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
import os

# LLM
llm = ChatGroq(
    model="openai/gpt-oss-20b",
    temperature=0.7,
    api_key=os.environ["GROQ_API_KEY"],
    verbose=True  # logs each LLM call to console
)

# --- Chain 1: Cuisine → Restaurant Name ---
prompt_res_name = PromptTemplate(
    input_variables=["cuisine"],
    template="I want to open a restaurant for {cuisine} food, suggest some names for it."
)

# --- Chain 2: Restaurant Name → Menu Items ---
prompt_menu = PromptTemplate(
    input_variables=["restaurant_name"],
    template="Suggest some menu items for a restaurant called {restaurant_name}."
)

# ✅ LCEL Sequential Chain with intermediate output preserved
chain = (
    RunnablePassthrough.assign(
        restaurant_name = prompt_res_name | llm | StrOutputParser()
    )
    | RunnablePassthrough.assign(
        menu_items = (
            lambda x: {"restaurant_name": x["restaurant_name"]}
        )
        | prompt_menu | llm | StrOutputParser()
    )
)

# Run it
result = chain.invoke({"cuisine": "Indian"})

```

```python
from IPython.display import display, Markdown

display(Markdown("### Restaurant Name"))
display(Markdown(result["restaurant_name"]))

display(Markdown("### Menu Items"))
display(Markdown(result["menu_items"]))
```

# Simple Chat Bot

```python
from langchain_groq import ChatGroq

llm = ChatGroq(model="openai/gpt-oss-20b", max_tokens = 2500)

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        break

    response = llm.invoke(user_input)
    print("Bot:", response.content)
```

```python

from langchain_community.agent_toolkits.load_tools import load_tools
from langgraph.prebuilt import create_react_agent          # ✅ correct location in 2025/26

# ── API Keys ──────────────────────────────────────────────────────────────────
from google.colab import userdata
api_key = userdata.get('SERP_API_KEY')

# ── LLM ───────────────────────────────────────────────────────────────────────
llm = ChatGroq(model="openai/gpt-oss-20b", temperature=0.9,  api_key = os.environ["GROQ_API_KEY"])

# ── Tools ─────────────────────────────────────────────────────────────────────
tools = load_tools(['serpapi'], llm=llm)

# ── Agent (LangGraph style) ───────────────────────────────────────────────────
# No separate AgentExecutor needed — LangGraph handles the loop internally
agent = create_react_agent(
    model=llm,
    tools=tools,
)

# ── Run ───────────────────────────────────────────────────────────────────────
result = agent.invoke({
    "messages": [{"role": "user", "content": "Who won the last mens cricket ODI World Cup?"}]
})

# ── Output ────────────────────────────────────────────────────────────────────
print("\n📢 Final Answer:", result['messages'][-1].content)
```
